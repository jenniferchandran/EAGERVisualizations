```python
from multiprocessing import Pool, cpu_count
import pandas
import geocoder
```

```python
cpu_count()
```

```python
data = pandas.read_excel('state_added.xlsx')
```

```python
pool = Pool(cpu_count()) 
```

```python
data
```

```python
args_list = [(data.iloc[r]['Latitude'], data.iloc[r]['Longitude']) for r in range(len(data))]
```

```python
geocoder.get_state_abbr_from_coords(args_list[0])
```

```python
results = []

for idx, res in enumerate(pool.imap(geocoder.get_state_abbr_from_coords, args_list), 1):
    results.append(res)
    print(f"Finished task {idx} of {len(args_list)}")
```

```python
data['state_added'] = results
```
